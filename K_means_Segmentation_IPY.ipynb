# **Step 2: Melakukan teknik Segmentasi K-means pada gambar sinar-x Covid-19 yang Ditingkatkan**

```
def KMeans():
  inPath ="drive/MyDrive/SkripsiData224/trainEnhanc/normal"
  outPath ="drive/MyDrive/SkripsiData244/trainKmeans/normal"
  for imagePath in os.listdir(inPath):
    inputPath=os.path.join(inPath, imagePath)
    img=Image.open(inputPath)
    numpy_image = np.array(img)
    opencvImage = cv2.cvtColor(numpy_image, cv2.COLOR_BGR2RGB)
    #ubah gambar MxNx3 menjadi matriks Kx3 di mana K=MxN dan setiap baris sekarang menjadi vektor dalam ruang 3-D RGB.
    pixel_values = opencvImage.reshape((-1, 3))
    #mengonversi nilai unit8 menjadi float karena ini merupakan persyaratan metode k-means dari OpenCV.
    pixel_values = np.float32(pixel_values)
    #hentikan iterasi algoritma jika epsilon tercapai atau entikan algoritma setelah max iterasi, max iterasi, nilai epsilon 
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
    #jumlah cluster
    k = 7
    _, labels, (centers) = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
    centers = np.uint8(centers)

    labels = labels.flatten()
    segmented_image = centers[labels.flatten()]
    segmented_image = segmented_image.reshape(opencvImage.shape)

    fullOutPath=os.path.join(outPath,'segmentedKMeans_'+imagePath)
    im_pil = Image.fromarray(segmented_image)
    im_pil.save(fullOutPath)
    print(fullOutPath) 
```

```
def main():
  KMeans()
if __name__=='__main__':
  main()
```
